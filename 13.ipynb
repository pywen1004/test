13. Perform experimental analysis to test the hypothesis that Pythonâ€™s sorted method runs in O(nlogn) time on average.

```python
import random
import time
import matplotlib.pyplot as plt
import numpy as np

def sorting():
    sizes = [10, 100, 1000, 10000, 100000, 1000000]
    times = []

    for size in sizes:
        random_list = [random.randint(0, 1000000) for _ in range(size)]

        start_time = time.time()
        sorted_list = sorted(random_list)
        end_time = time.time()

        elapsed_time = end_time - start_time
        times.append(elapsed_time)

        print(f"Size: {size}, Time taken: {elapsed_time:.6f} seconds")

    return sizes, times

def plot(sizes, times):
    n_log_n = [n * np.log(n) for n in sizes]

    plt.figure(figsize=(10, 6))
    plt.plot(sizes, times, label='Measured Time', marker='o')
    plt.plot(sizes, n_log_n, label='O(n log n)', linestyle='--')
    plt.xlabel('Input Size (n)')
    plt.ylabel('Time (seconds)')
    plt.title("Performance of Python's sorted() method")
    plt.legend()
    plt.grid()
    plt.show()

sizes, times = sorting()
plot(sizes, times)
```
